## 2. Split data into train and test sets

```python
# Save data for approximately the last 9 months
test_length = 9
# Remove the test data from the data set
train_length = len(df)- test_length
ts = df.values
ts_test = ts[-test_length:]
ts_train = ts[:train_length]
```

```python
print(ts_test.shape)
print(ts_train.shape)
```

## 3. Normalize data
Because the input data with wildly different mean might prevent model convergence, we should normalize this data to make sure when we apply machine learning algorithm, the model will be effected by the variance of the data. Normalizing data means we will going to rescale the data from the original range of all values to be within the range of 0 and 1.



```python
"""
Thanks to Sklearn we can use [MinMaxScaler]
(http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html) to normalize data
from sklearn.preprocessing import MinMaxScaler
# prepare data for normalization
values = ts
# train the normalization
scaler = MinMaxScaler(feature_range=(0, 1))
scaler = scaler.fit(values)
# normalize the dataset and print the first 5 rows
ts_normed = scaler.transform(values)
for i in range(5):
    print(ts_normed[i])
# inverse transform and print the first 5 rows
inversed = scaler.inverse_transform(ts_normed)
for i in range(5):
    print(inversed[i])
"""
means = ts[:train_length, :].mean(0)
stds = ts[:train_length, :].std(0)
ts_normed = (ts - means) / stds
assert ts_normed.shape == ts.shape and np.sum(np.isnan(ts_normed)) == 0
```

## 5. Build an LSTM model for your data


```python
#Utility functions, which are already written for you. However, make sure to read the code carefully and understand line by line

def compute_mape(model, x, y):
    forecasts = model.predict(x, batch_size=len(x))
    forecasts = (forecasts * stds) + means
    y_denom = (y * stds) + means
    return np.mean(np.abs((y_denom - forecasts) / y_denom))
    
def test(model):
    print('Train Mean Absolute Percentage Error (MAPE): {0:.3f}'.format(compute_mape(model, X_train, y_train)))
    print('Test Mean Absolute Percentage Error (MAPE): {0:.3f}'.format(compute_mape(model, X_test, y_test)))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import LSTM
from keras.callbacks import EarlyStopping 

dim = 26

# Hyper parameters
sequence_length = 20 # sequence is the number of previous time steps to use as input variables to predict the next time period
num_rnn_layers = 3
num_rnn_units = 64
batch_size = 8
epochs = 500

# Generate X and y for train/test
# Xs are 3D tensors of shape number of sequences * sequence length * dim
# ys are 2D tensors of shape number of sequences * dim
X_train, y_train, X_test, y_test = [], [], [], []

for i in range(train_length - sequence_length):
    X_train.append(ts_normed[i:i+sequence_length])
    y_train.append(ts_normed[i+sequence_length])

for i in range(-test_length, 0):
    y_test.append(ts_normed[i])
    X_test.append(ts_normed[i-sequence_length:i])

    
X_train, y_train = np.array(X_train), np.array(y_train)
X_test, y_test = np.array(X_test), np.array(y_test) 

# Train model
print('Creating Model')

model = Sequential()
for i in range(num_rnn_layers):
    model.add(LSTM(num_rnn_units, 
                   input_shape=(sequence_length, dim),
                   return_sequences=True if i < num_rnn_layers-1 else False))
model.add(Dense(dim, kernel_initializer='glorot_normal'))
model.compile(loss='mse', optimizer='adam', metrics=['mse'])

print('Training')
# Early stopping is used to break the training process when validation loss is not improved. 
early_stopping = EarlyStopping(monitor='val_loss', patience=20)

history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1, verbose=2, 
                    callbacks=[early_stopping], shuffle=True)
#print(history.history)

plt.figure(figsize=(15,6))
plt.subplot(211)
plt.plot(history.epoch, history.history['loss'], 'r') 
plt.ylim(0, 2)
plt.subplot(212)
plt.plot(history.epoch, history.history['mean_squared_error'], 'r')
plt.ylim(0, 2)

print('Min. training loss: {0:.3f}'.format(min(history.history['loss'])))

print('Testing')
test(model)
```

```python
plt.figure(figsize=(15,6))
plt.subplot(211)
plt.plot(history.epoch, history.history['loss'], 'r') 
plt.ylim(0, 2)
plt.subplot(212)
plt.plot(history.epoch, history.history['val_mean_squared_error'], 'r')
plt.ylim(0, 2)

print('Min. training loss: {0:.3f}'.format(min(history.history['loss'])))

print('Testing')
test(model)
```

### 5.2. Analyze results and play with different parameters to get a sense of how they affect the performance.
